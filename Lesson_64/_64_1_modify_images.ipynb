За допомогою графічного редактора, визначте точки відповідності (кути дошки) на зображеннях chess_1 та chess_2, що додні до репозиторію лекції. Проведіть гомографічне преретворення для приведення зображення chess_2 до вигляду зображення chess_1.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_src = cv2.imread('chess_2.png')
pts_src = np.array([[215, 300],
                    [220, 80],
                    [380, 350],
                    [460, 120]])
plt.figure(figsize=(12,8))
plt.imshow(im_src)

plt.show()
```

```python
im_dst = cv2.imread('chess_1.png')
pts_dst = np.array([[230, 80],
                    [430, 80],
                    [230, 220],
                    [410, 230]])
plt.figure(figsize=(12,8))
plt.imshow(im_dst)

plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst,)
```

```python
display(h)
```

```python
im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1], im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)

plt.show()
```
