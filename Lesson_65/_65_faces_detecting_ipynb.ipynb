```python
import dlib
from time import time
import matplotlib.pyplot as plt
from scipy import ndimage
import cv2
from imutils import face_utils
```

**1. Алгоритм Віола-Джонса для розпізнавання облич.**

```python
face_cascade = cv2.CascadeClassifier("/content/drive/MyDrive/Resume/lection_65.xml")  
```

```python
image = cv2.imread("/content/drive/MyDrive/Resume/titanic.jpg")                        

image_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)                                  
plt.imshow(image_color)
plt.axis('off')
plt.show    
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)                                                     
face_found = face_cascade.detectMultiScale(image_gray, 1.3, 5)
for (x, y, w, h) in face_found:
    cv2.rectangle(image_color, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(image_color)
plt.axis('off')
plt.show
```

```python
image_rotate = cv2.rotate(image_color, cv2.ROTATE_90_CLOCKWISE)                  
image_gray = cv2.cvtColor(image_rotate, cv2.COLOR_BGR2GRAY)
face_found = face_cascade.detectMultiScale(image_gray, 1.3, 5)
for (x, y, w, h) in face_found:
    cv2.rectangle(image_rotate, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(image_rotate)
plt.axis('off')
plt.show;
```
