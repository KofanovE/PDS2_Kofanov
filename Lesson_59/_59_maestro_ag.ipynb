```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('maestro_ag.jpg')
img = cv2.resize(img, (625, 1000))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
def sobel_filter(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1],
                        [0, 0, 0],
                        [1, 2, 1]])
    image_x = cv2.filter2D(gray_image, -1, sobel_x)
    image_y = cv2.filter2D(gray_image, -1, sobel_y)
    combined_image = cv2.bitwise_or(image_x, image_y)
    return combined_image
```

```python
sobel_filter_image = sobel_filter(img)
fig = plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(sobel_filter_image)
plt.show()
```

```python
def median_filter(image, kernel_size):
    filtered_image = cv2.medianBlur(image, kernel_size)
    return filtered_image
```

```python
median_filter_image = median_filter(img, kernel_size=5)
fig = plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(median_filter_image)
plt.show()
```

```python
plt.subplot(1,3,1),plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobel_filter_image, cmap = 'gray')
plt.title('Sobel Filter'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(median_filter_image, cmap = 'gray')
plt.title('Median Filter'), plt.xticks([]), plt.yticks([])
plt.show()
```

**В цілому, фільтр Собеля і медіанний фільтр - два різних метода обробки зображень. Фільтр Собеля використовується для виділення контурів, у той час, як медіанний фільтр використовується для видалення шумів і зменшення деталей, що чітко і простежується на даному прикладі.**
