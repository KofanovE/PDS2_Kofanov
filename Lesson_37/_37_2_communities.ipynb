Використовуючи датасет з файлу communities.csv, виконайте нормалізацію даних. Визначте найкращий метод для нормалізації даних даного набору даних.

**1. Завантаження датасету та відображення декількох рядків датасету для зразку.**

```python
import pandas as pd

df = pd.read_csv('communities.csv')
df.head(10)
```

**2. Замінимо всі данні зі знаком питання на Nan.**

```python
import numpy as np

df = df.replace('?', np.nan)
df
```

**3. Видалимо всі колонки, де зустрічаються Nan**

```python
df = df.dropna(axis=1, how='any')
df
```

**4. Видалимо колонки з нечисловими даними.**

```python
df = df.drop(columns=df.select_dtypes(exclude=['float64', 'int64']).columns)
df
```

**5. Визначимо останній стовпчик датасету як цільовий та знайдемо найбільш зкорельовану з ним частину датасету.**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn import utils

lab = preprocessing.LabelEncoder()

X = df.drop(df.columns[-1:], axis=1)
y = lab.fit_transform(df.iloc[:, -1:])

rf = RandomForestClassifier()
rf.fit(X, y)
importances = rf.feature_importances_

importances
```

***Робимо висновок, що датасет не має зкорельованих ознак із цільовими даними, тому будемо використовувати всі ознаки.***

**6. Сформуємо тестовий та тренувальний датасети та побудуємо модель множинної регресії.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=20)

lr = LinearRegression()
lr.fit(X_train, y_train)

print("Якість моделі лінійної регресії на тестовому наборі: {:.4f}".format(lr.score(X_test, y_test)))
```

**7. Проведемо L1 регулярізацію моделі.**

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.001).fit(X_train, y_train)

print("Якість моделі з регулярізацією L1 на тестовому наборі: {:.4f}".format(lasso.score(X_test, y_test)))
```

**8. Проведемо L2 регулярізацію моделі.**

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1).fit(X_train, y_train)

print("Якість моделі з регулярізацією L2 на тестовому наборі: {:.4f}".format(ridge.score(X_test, y_test)))
```

**9. Проведемо L1 + L2 регулярізацію моделі.**

```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(alpha=0.0001, l1_ratio=.1).fit(X_train, y_train)

print("Якість моделі з регулярізацією L1 + L2 на тестовому наборі: {:.4f}".format(elastic.score(X_test, y_test)))
```

***Зміни параметрів всіх типів регулярізації на якість моделі лінійної регресії не вплинули; всі моделі виявилися досить низької якості. На мою думку, це перед усім пояснюється самим характером даних.***
