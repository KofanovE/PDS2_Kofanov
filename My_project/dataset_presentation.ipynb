```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
parent_dir = 'C:/Users/админ/PycharmProjects/PDS2/PDS2_Kofanov/My_project/Datasets/dataset_parser'
data = []
```

```python
for dir_name in os.listdir(parent_dir):
    if os.path.isdir(os.path.join(parent_dir, dir_name)):
        files_count = len(os.listdir(os.path.join(parent_dir, dir_name)))
        data.append([dir_name, files_count])
```

```python
df = pd.DataFrame(data, columns=['class', 'images_number'])
```

```python
vc = df["class"].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x = vc.index, y = df["images_number"], palette = "rocket")
plt.title("Number of pictures of each category", fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(15, 15), subplot_kw={'xticks': [], 'yticks': []})
subdirs = [d for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]
for i, subdir in enumerate(subdirs):
    images = [f for f in os.listdir(os.path.join(parent_dir, subdir)) if os.path.isfile(os.path.join(parent_dir, subdir, f))]
    axes[i//5, i%5].imshow(plt.imread(os.path.join(parent_dir, subdir, images[0])))
    axes[i//5, i%5].set_title(subdir)
plt.show()
```
